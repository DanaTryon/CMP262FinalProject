I am not interested in all of the columns that are 7 day averages so the following cell will remove them.

```python
# Define the columns to remove
columns_to_remove = list(range(5, 9)) + list(range(13, 18)) + list(range(22, 27)) + \
                    list(range(31, 36)) + list(range(40, 45)) + list(range(49, 55))

# Remove the specified columns
df.drop(df.columns[columns_to_remove], axis=1, inplace=True)

# Print the DataFrame to verify the columns are removed
df.info()
```

The following cell will remove the underscore character from the column names and replace with a space for better readability.

```python
# Replace underscores with space strings in column names
df.columns = df.columns.str.replace('_', ' ')

# Print the updated DataFrame to verify the changes
df.info()
```

Now to save the cleaned dataframe to a renamed csv file.

```python
# Save the DataFrame as a CSV file
df.to_csv(r'C:\Users\Danat\OneDrive\Documents\CMP-262\FinalProject\Cleaned-NYC-COVID-19.csv', index=False)
```
