### Question 4.  Is there a particular time of the year with the most impact due to Covid-19?
Usually the winter months are the worst for any type of respiratory illness that is spread from exhaled viral substances.  By grouping the total hospitalizations by month, we can see what times of the year the hospitals will be the busiest.

```python
import seaborn as sns

# Assuming we have a DataFrame 'df' with a column 'date of interest'
# We will create a new column 'month' that contains the string value of the month

# Define a dictionary to map month numbers to month names
month_mapping = {
    '01': 'January',
    '02': 'February',
    '03': 'March',
    '04': 'April',
    '05': 'May',
    '06': 'June',
    '07': 'July',
    '08': 'August',
    '09': 'September',
    '10': 'October',
    '11': 'November',
    '12': 'December'
}

# Extract the first two digits from the 'date of interest' column
df['month'] = df['date of interest'].apply(lambda x: month_mapping[x.split('/')[0]])


hosp_counts = df[['HOSPITALIZED COUNT', 'month']]  #creating a new data frame with only 2 columns to make graphing easier
monthly_totals = hosp_counts.groupby('month')['HOSPITALIZED COUNT'].sum().reset_index()  #summing total counts by month
#need to make a list of the months in their natural order so that they do not get displayed by alphabetical order on the x axis
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
monthly_totals['month'] = pd.Categorical(monthly_totals['month'], categories=month_order, ordered=True)
monthly_totals.sort_values('month', inplace=True)
print('Total NYC hospitalizations for the entire duration of the pandemic is: ' + f"{df['HOSPITALIZED COUNT'].sum()}")
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size

sns.barplot(x='month', y='HOSPITALIZED COUNT', data=monthly_totals, palette='viridis')
plt.title('Monthly Sum of Hospitalized Counts in NYC for the duration of the Covid-19 Pandemic')
plt.xlabel('Month')
plt.ylabel('Total Hospitalized Count')

plt.show()  # Display the plot
```

January was the peak month for hospitalizations with December through April being the busy time of the year for hosptials.  There is a 7 month reprieve from May through November.  Hospital administrators are probably very reluctant to approve vacation time for staff for the December/January holiday season based on this chart.

The following graph will demonstrate some of the very bad days as outliers by plotting each single day of the pandemic by daily case count on the x axis and daily hospitalization count on the y axis.  Each point shows a day in which there were confirmed Covid-19 cases and how many of these cases required hospitalization.  These points are grouped by each month as a different color.

```python
import seaborn as sns
sns.set_style('whitegrid')
ax = sns.scatterplot(data=df, x= df['CASE COUNT'], y=df['HOSPITALIZED COUNT'], hue='month')
ax.set_title('Daily Case Counts v Daily Hospitalization Counts', fontdict={'size': 15, 'weight': 'bold'})
# Move the legend to the upper left corner outside of the plot
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
# Show the plot
plt.show()
```

One interesting part of this graph is the line of dark orange and olive dots going up the y axis along the 5,000 case count x axis line.  This is corresponding to all of the hospitalizations that occured during the first wave in the months of March and April.  Another interesting part of the graph is the line of pink and light purple dots going up a 45 degree angle from the origin.  These extreme case count numbers are from the third wave that occured in December of 2021 and January of 2022.  Using this graph as a 4 quadrant analysis, the months of December and January produced quite a few upper right quadrant points that demonstrate these are the worst months for getting Covid-19 and being hospitalized for it.
